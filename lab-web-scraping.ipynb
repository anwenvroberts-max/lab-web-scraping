```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Map str descriptions "One", "Two".. to int numbering 1, 2...
def get_rating_number(rating_classes):
    mapping = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    for word, number in mapping.items():
        if word in rating_classes:
            return number
    return 0
```

```python
## Put URL call IN the Fct. scrape_books 

def scrape_books(min_rating, max_price):    # make rating and price constrictions Variables of the scrape_books function
    base_url = "https://books.toscrape.com/catalogue/"
   
    page_url = base_url + "page-1.html"      # Scrape first page only 
    response = requests.get(page_url)      # Get HTML catalog page with BSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    books_data = []
    books = soup.find_all('article', class_='product_pod') # Find book like items on page
    
    for book in books:
        price_text = book.find('p', class_='price_color').text  # Get price and clean - drop £ signs etc. to process as int
        price = float(price_text.replace('£', '').replace('Â', ''))
        
        rating_tags = book.find('p', class_='star-rating')['class']     # Inspect CSS class names with nested dict
        rating = get_rating_number(rating_tags)
        
        if rating >= min_rating and price <= max_price:       # Filter books according to rating and price criteria - pass as Var in fct call! 
            
            book_link = book.h3.a['href']    # Get URL to lookup "Book page" 
            detail_response = requests.get(base_url + book_link)
            detail_soup = BeautifulSoup(detail_response.text, 'html.parser')
                  
            title = book.h3.a['title']    # Extract data from detail page 
            upc = detail_soup.find('th', string='UPC').find_next_sibling('td').text
            genre = detail_soup.find('ul', class_='breadcrumb').find_all('li')[2].text.strip()
            availability = detail_soup.find('p', class_='instock availability').text.strip()
            
            # Instead of empty "class" get the header above and the respective next paragraph
            desc_header = detail_soup.find('div', id='product_description')
            description = desc_header.find_next_sibling('p').text if desc_header else "No description"
            
            # Store each Book entry in a dict
            books_data.append({
                'UPC': upc,
                'Title': title,
                'Price (£)': price,
                'Rating': rating,
                'Genre': genre,
                'Availability': availability,
                'Description': description
            })
   
    # Convert list of dict to pandas D.F.
    df = pd.DataFrame(books_data)
    return df
```

```python
## Scrape ONLY books with a min_rating =4 and max_price=20

# Pass to function with specs: minimum rating 4, and max price 20
results_df = scrape_books(min_rating=4, max_price=20.0)
print(results_df)
```
